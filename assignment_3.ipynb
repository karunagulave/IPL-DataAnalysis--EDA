```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
matches_df = pd.read_csv("/content/IPL Matches 2008-2020.csv")
balls_df = pd.read_csv("/content/IPL Ball-by-Ball 2008-2020.csv")
```

```python
matches_df
```

```python
balls_df
```

```python
matches_df['date']=pd.to_datetime(matches_df['date'])
```

```python
matches_df.info()
```

```python
matches_df['season']=matches_df['date'].dt.year
```

```python
matches_df.info()
```

```python
#1. what was the count of matches played in each season.
#Answer:-
matches_per_season=matches_df['season'].value_counts().sort_index()
plt.figure()

sns.barplot(x=matches_per_season.index, y=matches_per_season.values,color="blue")
plt.title("Matches played in each season")

plt.xlabel("Season")
plt.ylabel("Count")

plt.tight_layout()

plt.show()
```

```python
# 2. How many runs were scored in each season?
#Answer:-
merged_df = balls_df.merge(matches_df[["id","season"]],on='id')
runs_per_season = merged_df.groupby('season')['total_runs'].sum()

plt.figure()
sns.barplot(x=runs_per_season.index, y=runs_per_season.values,palette="viridis")
plt.title("Runs scored in each season")

plt.xlabel("Season")
plt.ylabel(" Total Runs")

plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

```python
#3. What were the runs scored per match in different seasons?
#Answer:-
runs_per_match_season = runs_per_season / matches_per_season
runs_per_match_season
```

```python
# 4. Who has umpired the most?
#Answer:-
umpire_counts = matches_df['umpire1'].value_counts() + matches_df['umpire2'].value_counts()
most_umpired_umpire = umpire_counts.idxmax()
most_umpired_umpire
```

```python
# 5. Which team has won the most tosses?
#Answer:-
most_toss_winners = matches_df['toss_winner'].value_counts().idxmax()
most_toss_winners
```

```python
# 6. What does the team decide after winning the toss?
#Answer:-
toss_decision_counts= matches_df['toss_decision'].value_counts()
toss_decision_counts
```

```python
# 7. How does the toss decision vary across seasons?
#Answer:-
toss_decision_season = matches_df.groupby('season')['toss_decision'].value_counts()
toss_decision_season
```

```python
# 8. Does winning the toss imply winning the game?
#Answer:-
toss_win_counts = matches_df.groupby('toss_winner')['winner'].value_counts()
toss_win_counts
```

```python
# 9. How many times has the chasing team won the match?
#Answer:-
chasing_team_wins = (matches_df['team1'] != matches_df['winner']).sum() + (matches_df['team2'] != matches_df['winner']).sum()
chasing_team_wins
```

```python
#10. Which all teams had won this tournament?
#Answer:-
winning_teams = matches_df['winner'].unique()
winning_teams
```

```python
#11. Which team has played the most number of matches?
#Answer:-
most_played_team = matches_df['team1'].value_counts().idxmax()
most_played_team
```

```python
#12. Which team has won the most number of times?
#Answer:-
most_winning_team = matches_df['winner'].value_counts().idxmax()
most_winning_team
```

```python
#13. Which team has the highest winning percentage?
#Answer:-
winning_percentage = matches_df['winner'].value_counts(normalize=True) * 100
winning_percentage
```

```python
#14. Is there any lucky venue for a particular team?
#Answer:-
venue_counts = matches_df['venue'].value_counts()
venue_counts
```

```python
#15. Innings wise comparison between teams
#Answer:-
comparison_df=merged_df.groupby(['season','inning','batting_team'])['total_runs'].sum().reset_index()
comparison_df=comparison_df.pivot_table(index=['season','inning'],columns='batting_team',values='total_runs',fill_value=0)
comparison_df
```

```python
#16. Which team has scored the most number of 200+ scores?
#Answer:-
top_scorers = balls_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
top_scorers
```

```python
#17. Which team has conceded 200+ scores the most?
#Answer:-
top_concen = balls_df.groupby('bowler')['total_runs'].sum().sort_values(ascending=False)
top_concen
```

```python
#18. What was the highest run scored by a team in a single match?
#Answer:-
highest_runs_per_match = balls_df.groupby('id')['total_runs'].sum().max()
highest_runs_per_match
```

```python
#19. Which is the biggest win in terms of run margin?
#Answer:-
biggest_win_df=merged_df[merged_df['total_runs']==merged_df['total_runs'].max()]
biggest_win_df=biggest_win_df[['season','id','batting_team','bowling_team','total_runs']]
biggest_win_df.head(1)
```

```python
#20. Which batsmen have played the most number of balls?
#Answer:-
balls_faced = balls_df.groupby('batsman')['ball'].count()
balls_faced
```

```python
#21. Who are the leading run-scorers of all time?
#Answer:-
top_run_scorers = balls_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
top_run_scorers
```

```python
#22. Who has hit the most number of 4's?
#Answer:-
four_hitters = balls_df[balls_df['batsman_runs'] == 4].groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
four_hitters
```

```python
#23. Who has hit the most number of 6's?
#Answer:-
six_hitters = balls_df[balls_df['batsman_runs'] == 6].groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
six_hitters
```

```python
# 24. Who has the highest strike rate?
#Answer:-
strike_rate = balls_df.groupby('batsman')['batsman_runs'].count() / balls_df.groupby('batsman')['ball'].count() * 100
strike_rate
```

```python
#25. Who is the leading wicket-taker?
#Answer:-
wicket_takers = balls_df[balls_df['player_dismissed'].notnull()].groupby('player_dismissed')['player_dismissed'].count().sort_values(ascending=False)
wicket_takers
```

```python
#26. Which stadium has hosted the most number of matches?
#Answer:-
most_hosted_stadium = matches_df['venue'].value_counts().idxmax()
most_hosted_stadium

```

```python
#27. Who has won the most MOM awards?
#Answer:-
most_mom_awards = matches_df['player_of_match'].value_counts().idxmax()
most_mom_awards
```

```python
#28. What is the count of fours hit in each season?
#Answer:-
fours_per_season=merged_df[merged_df['batsman_runs']==4]
fours_per_season=fours_per_season.groupby('season')['batsman_runs'].count()
fours_per_season=fours_per_season.sort_values(ascending=False)
fours_per_season.head(1)
```

```python
#29. What is the count of sixes hit in each season?y
#Answer:-
sixes_per_season=merged_df[merged_df['batsman_runs']==6]
sixes_per_season=sixes_per_season.groupby('season')['batsman_runs'].count()
sixes_per_season
```

```python
#30. What is the count of runs scored from boundaries in each season?
#Answer:-
boundaries_per_season=merged_df[(merged_df['batsman_runs']==4) | (merged_df['batsman_runs']==6)]
boundaries_per_season=boundaries_per_season.groupby('season')['batsman_runs'].count()
boundaries_per_season
```

```python
#31. What is the run contribution from boundaries in each season?
#Answer:-
boundaries_per_season=merged_df[(merged_df['batsman_runs']==4) | (merged_df['batsman_runs']==6)]
boundaries_per_season=boundaries_per_season.groupby('season')['batsman_runs'].count()
boundaries_per_season=boundaries_per_season/merged_df.groupby('season')['batsman_runs'].count()*100
boundaries_per_season
```

```python
# 32. Which team has scored the most runs in the first 6 overs?
#Answer:-
first_6_overs_runs = balls_df[balls_df['over'] <= 6].groupby('batting_team')['total_runs'].sum().idxmax()
first_6_overs_runs
```

```python
# 33. Which team has scored the most runs in the last 4 overs?
#Answer:-
last_4_overs_runs = balls_df[balls_df['over'] >= 16].groupby('batting_team')['total_runs'].sum().idxmax()
last_4_overs_runs
```

```python
# 34. Which team has the best scoring run-rate in the first 6 overs?
#Answer:-
first_6_overs_run_rate = balls_df[balls_df['over'] <= 6].groupby('batting_team')['total_runs'].sum() / balls_df[balls_df['over'] <= 6].groupby('batting_team')['ball'].count()
first_6_overs_run_rate
```

```python
# 35. Which team has the best scoring run-rate in the last 4 overs?
#Answer:-
last_4_overs_run_rate = balls_df[balls_df['over'] >= 16].groupby('batting_team')['total_runs'].sum() / balls_df[balls_df['over'] >= 16].groupby('batting_team')['ball'].count()
last_4_overs_run_rate
```
